## 2. Atendimentos com valor entre R$90 e R$200

```python
df_90_200 = pd.DataFrame(list(db.atendimentos.find({"valor": {"$gte": 90, "$lte": 200}}, {"_id": 0})))
print(f'{len(df_90_200)} atendimentos entre R$90 e R$200')
df_90_200
```

## 3. Atendimentos das cidades Assis, Ourinhos, Pompeia e Marília

```python
df_cidades = pd.DataFrame(list(db.atendimentos.find({"cidade": {"$in": ["Assis", "Ourinhos", "Pompeia", "Marilia"]}}, {"_id": 0})))
print(f'{len(df_cidades)} atendimentos encontrados')
df_cidades
```

## 4. Converter resultados para DataFrame

```python
df = pd.DataFrame(list(db.atendimentos.find({}, {"_id": 0})))
print(f'DataFrame com {len(df)} registros e {len(df.columns)} colunas')
df.head()
```

## 5. Verificar valores nulos

```python
print(df.isnull().sum())
```

## 6. Faturamento total

```python
faturamento_total = df['valor'].sum()
print(f'Faturamento total: R${faturamento_total:.2f}')
```

## 7. Faturamento por cidade

```python
faturamento_por_cidade = df.groupby('cidade')['valor'].sum().to_dict()
print(faturamento_por_cidade)
```

## 8. Faturamento por serviço

```python
faturamento_por_servico = df.groupby('servico')['valor'].sum().to_dict()
print(faturamento_por_servico)
```

## 9. Ticket médio geral

```python
ticket_medio_geral = df['valor'].mean()
print(f'Ticket médio geral: R${ticket_medio_geral:.2f}')
```

## 10. Ticket médio por cidade

```python
ticket_medio_por_cidade = df.groupby('cidade')['valor'].mean().to_dict()
print(ticket_medio_por_cidade)
```

## 11. Serviço mais lucrativo

```python
servico_mais_lucrativo = max(faturamento_por_servico, key=faturamento_por_servico.get)
print(f'Serviço mais lucrativo: {servico_mais_lucrativo} (R${faturamento_por_servico[servico_mais_lucrativo]:.2f})')
```

## 12. Cidade com maior faturamento

```python
cidade_maior_faturamento = max(faturamento_por_cidade, key=faturamento_por_cidade.get)
print(f'Cidade com maior faturamento: {cidade_maior_faturamento} (R${faturamento_por_cidade[cidade_maior_faturamento]:.2f})')
```

## 13. Top 3 maiores atendimentos

```python
top3 = df.nlargest(3, 'valor')[['tutor', 'nome_pet', 'servico', 'valor', 'cidade']]
print(top3)
```

## Relatório Final — Geração e Salvamento no MongoDB

```python
relatorio = {
    "data_geracao": datetime.now(),
    "faturamento_total": float(faturamento_total),
    "faturamento_por_cidade": {k: float(v) for k, v in faturamento_por_cidade.items()},
    "faturamento_por_servico": {k: float(v) for k, v in faturamento_por_servico.items()},
    "ticket_medio_geral": float(ticket_medio_geral),
    "ticket_medio_por_cidade": {k: float(v) for k, v in ticket_medio_por_cidade.items()},
    "servico_mais_lucrativo": servico_mais_lucrativo,
    "cidade_maior_faturamento": cidade_maior_faturamento,
    "top_3_atendimentos": df.nlargest(3, 'valor').drop(columns=['data']).to_dict(orient='records')
}

db.relatorios.insert_one(relatorio)
relatorio.pop('_id', None)
print(relatorio)
```
